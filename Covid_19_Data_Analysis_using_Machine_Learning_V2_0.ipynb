```python
#Linear regression
from sklearn.model_selection import train_test_split
```

```python
maha
```

```python
#converting date-time to ordinal
maha['date']=maha['date'].map(dt.datetime.toordinal)
maha.head()
```

```python
#getting dependent variable and inpedent variable
x=maha['date']
y=maha['confirmed']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
maha.tail()
```

```python
lr.predict(np.array([[737630]]))
```

```python
# Create an array of dates for which you want to make predictions
dates_to_predict = np.array([[737630], [737631], [737632], [737633]])
```

```python
# Convert the dates to readable datetime format
dates = [dt.datetime.fromordinal(int(date_ordinal)) for date_ordinal in dates_to_predict]
```

```python
# Predict the number of confirmed cases for the specified dates
predicted_cases = lr.predict(dates_to_predict.reshape(-1, 1))
```

```python
# Create a DataFrame to hold the predicted data
predicted_data = pd.DataFrame({'date': dates, 'confirmed': predicted_cases.flatten()})
```

```python
# Convert the 'date' column to ordinal for plotting
predicted_data['date'] = predicted_data['date'].map(dt.datetime.toordinal)
```

```python
# Plotting the predicted data
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date", y="confirmed", data=predicted_data, marker='o', color="b", label="Predicted Cases")
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Predicted Number of Confirmed COVID-19 Cases in Maharashtra')
plt.legend()
plt.show()
```
